## 2.1 Ordinal/Categorical variables

### 2.1.1 Are the loans with good credit grades?

```python
df.CreditGrade.sort_values().value_counts()
```

The credit grades are not really good, the marjority of loans have Grade C and D. To better understand, let's plot a graph with percentage.

```python
base_color = sb.color_palette()[2]
```

```python
# convert the data into an ordered type, so the credit grades are ordered innately
level_order = ['AA', 'A', 'B', 'C', 'D', 'E', 'HR', 'NC']
ordered_credit = pd.api.types.CategoricalDtype(ordered = True, categories = level_order)
df['CreditGrade'] = df['CreditGrade'].astype(ordered_credit)

sb.countplot(df.CreditGrade, color = base_color)
plt.title('Distribution of Credit Grades')
plt.ylabel('No of borrowers')
plt.xlabel('Credit Grades');
```

**Observations:**

Now it is clearer that the majority of borrowers receive C and D grades credits. Next, I want to see if a pie chart showing proportion here would facilitate understanding.

```python
# display both % and absolute values
def func(pct, allvals):
    absolute = int(pct/100.*np.sum(allvals))
    return "{:.1f}%\n({:d})".format(pct, absolute)
```

```python
# define data and labels for the pie chart 
data=df.CreditGrade.sort_values().value_counts()
labels=data.index

# plot
plt.pie(data,
        autopct=lambda pct: func(pct, data),
       labels=labels, startangle = 90,
        counterclock = False,
       wedgeprops = {'width' : 0.4})
plt.axis('square') # so the scaling of the plot is equal on both the x- and y-axes
plt.title('Credit Grade Distribution');
```

**Observations:**

Although proportion is shown with regards to the whole, a pie chart seem to diminish the visual impact of the difference between credit grades. From empirical studies we know that it is more challenging for human to compare area of pie slices than the length of bars, so bar chart with percentage annotation might be a better choice here.

**Next steps:**
I will explore other rating variables with this learning in mind.

### 2.1.2 How about their Prosper Rating?

```python
# convert the data into an ordered type, so the ratings are ordered innately
level_order = ['AA', 'A', 'B', 'C', 'D', 'E', 'HR']
ordered_rating = pd.api.types.CategoricalDtype(ordered = True, categories = level_order)
df['ProsperRating (Alpha)'] = df['ProsperRating (Alpha)'].astype(ordered_rating)


ax = sb.countplot(df['ProsperRating (Alpha)'], color=base_color)
plt.title('Distribution of ProsperRating ')
plt.ylabel('No of borrowers')
plt.xlabel('Ratings')

# # set percentage annotation https://datascience.stackexchange.com/questions/48035/how-to-show-percentage-text-next-to-the-horizontal-bars-in-matplotlib
# total = len(df['ProsperRating (Alpha)'])
# for p in ax.patches:
#         percentage = '{:.1f}%'.format(100 * p.get_height()/total)
#         x = p.get_x() + p.get_width() - 0.6
#         y = p.get_y() + p.get_height()
#         ax.annotate(percentage, (x, y))

# add text for showing proportion 
rating_counts = df['ProsperRating (Alpha)'].value_counts()
# for i in range rating_counts[]
for i in range(rating_counts.shape[0]):
    count = rating_counts[i]
    pct_string = '{:0.1f}%'.format(100*count/rating_counts.sum())
    plt.text(i, count+2, pct_string, ha = 'center');
```

### 2.1.3 How long are the loan periods?

```python
ax = sb.countplot(df.Term)
# plt.figure(figsize=(15,5))
plt.title('Distribution of Loan Periods')
plt.ylabel('No of loan')
plt.xlabel('Term (Month)')

# # set percentage annotation https://datascience.stackexchange.com/questions/48035/how-to-show-percentage-text-next-to-the-horizontal-bars-in-matplotlib
# total = len(df['Term'])
# for p in ax.patches:
#         percentage = '{:.1f}%'.format(100 * p.get_height()/total)
#         x = p.get_x() + p.get_width() - 0.6
#         y = p.get_y() + p.get_height()
#         ax.annotate(percentage, (x, y))

# get total no of observation
n_points = df.shape[0] 
# calculate count by terms
term_count = df.Term.value_counts()
# get the current tick locations and labels
locs, labels = plt.xticks()

# loop through each pair of locations and labels
for loc, label in zip(locs, labels):
    
    # get the text property for the label to get the correct count
    # turn it to integer
    count = term_count[int(label.get_text())]
    pct_string = '{:0.1f}%'.format(100*count/n_points)
    plt.text(loc, count, pct_string, ha = 'center', color = 'b')
```

```python
df.Term.value_counts()
```

**Observations:**
Seems like 3 years is a common loan period. Most of the loans last for 3 years, with 87778 loans. The second is 5 years, with 24545 loans. Very few loans (1614) last for 1 year only.

```python
# all variables investigated above
all_var=[
    'CreditGrade',
    'ProsperRating (Alpha)',
    
    'LoanStatus',

    'TotalProsperPaymentsBilled',
    'ProsperPrincipalOutstanding',
    'ScorexChangeAtTimeOfListing',
    'LoanFirstDefaultedCycleNumber',

    'TotalProsperLoans',
    'PercentFunded',
    'Recommendations',
    'Investors']
```

```python
categoric_vars = ['CreditGrade','ProsperRating (Alpha)','LoanStatus']

numeric_vars = [
    'TotalProsperPaymentsBilled',
    'ProsperPrincipalOutstanding',
    'ScorexChangeAtTimeOfListing',
    'LoanFirstDefaultedCycleNumber',
    'TotalProsperLoans',
    'PercentFunded',
    'Recommendations',
    'Investors'
]

```

```python
# correlation coefficient heatmap
plt.figure(figsize = [8, 5])
sb.heatmap(df[numeric_vars].corr(), annot = True, fmt = '.3f',
           cmap = 'vlag_r', center = 0)
plt.show()
```

```python
# plot matrix: sample 500 diamonds so that plots are clearer and
# they render faster
samples = np.random.choice(df.shape[0], 500, replace = False)
df_samp = df.loc[samples,:]

g = sb.PairGrid(data = df_samp, vars = numeric_vars)
g = g.map_diag(plt.hist, bins = 100);
g.map_offdiag(plt.scatter);
```

**Observations:**

It appears that there are no strong relationship between the numeric variables explored. The strongly correlated ones are total payment billed and total loans - they have a positive relationship. Next we will look more into categorical variables.

```python
samples = np.random.choice(df.shape[0], 2000, replace = False)
df_samp = df.loc[samples,:]

def boxgrid(x, y, **kwargs):
    """ Quick hack for creating box plots with seaborn's PairGrid. """
    default_color = sb.color_palette()[0]
    sb.boxplot(x, y, color = default_color)

plt.figure(figsize = [10, 10])
g = sb.PairGrid(data = df_samp, y_vars = ['Investors', 'TotalProsperLoans'], x_vars = categoric_vars,
                size = 3, aspect = 1.5)
g.map(boxgrid)
plt.show();
```

```python
# to enlarge the plot
default_color = sb.color_palette()[0]
plt.figure(figsize=(25,10))
sb.boxplot(x=df.LoanStatus, y=df.TotalProsperLoans, color = default_color);
```

As we expected, it appears that the borrowers with higher credit grade seem to be supported by more investors, those with lower credit grades have less investors. Similar phenomenon can be observed in the Prosper Rating system as well. 

After enlarging one of the plot above, interestingly, we can see there are a few outlying borrowers with a higher number of total Prosper Loans when the loan status is completed, current, defaulted and charged off.
